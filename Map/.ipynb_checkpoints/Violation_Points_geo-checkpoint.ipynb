```python
import numpy as np
import pandas as pd
import shapefile as shp
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import plotly.express as px
from geopy.geocoders import Nominatim
from matplotlib.animation import FuncAnimation, PillowWriter
import calendar
```

```python
df = pd.read_csv('Clean_Data.csv')
df.drop("Unnamed: 0",axis =1,inplace=True)
df['Month'] = df['Month'].apply(lambda x: calendar.month_abbr[x])
df["City"] = df["City"].str.lower()
```

```python
shape = gpd.read_file("kx-king-county-wa-cities-and-unincorporated-areas-SHP/king-county-wa-cities-and-unincorporated-areas.shp")
shape["NAME"] = shape["NAME"].str.lower()
shape["NAME"].unique()
shape = shape.drop(labels=["UNINC"],axis = 1)
shape.boundary.plot()
```

```python
Years = df["Year"].unique().tolist()
Seasons = df["Season"].unique().tolist()
Months = df["Month"].unique().tolist()
day_of_weeks = df["day_of_week"].unique().tolist()
```

```python
df_Violation_Points = df[['Name', 'City', 'Violation_Points']].copy()
df_Violation_Points = df_Violation_Points.groupby('City').mean('Violation_Points').round()
shape_Violation_Points = pd.merge(left=shape, right=df_Violation_Points, left_on="NAME", right_on="City", how = 'inner')
shape_Violation_Points = gpd.GeoDataFrame(shape_Violation_Points)
ax = shape_Violation_Points.boundary.plot(edgecolor = 'black', linewidth = 0.2, figsize = (8,8))
shape_Violation_Points.plot(ax = ax , column ='Violation_Points',legend = True, cmap = "GnBu", legend_kwds={'shrink': 0.3})
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.set_title('Average Violation Points in each city', size = 15, weight = 'bold')
```

```python
for year in Years:
    df_Violation_Points = df[['Name', 'City', 'Violation_Points', 'Year']].copy()
    df_Violation_Points = df_Violation_Points[df_Violation_Points['Year'] == year]
    df_Violation_Points = df_Violation_Points.groupby('City').mean('Violation_Points').round()
    shape_Violation_Points = pd.merge(left=shape, right=df_Violation_Points, left_on="NAME", right_on="City", how = 'inner')
    shape_Violation_Points = gpd.GeoDataFrame(shape_Violation_Points)
    ax = shape_Violation_Points.boundary.plot(edgecolor = 'black', linewidth = 0.2, figsize = (8,8))
    shape_Violation_Points.plot(ax = ax , column ='Violation_Points',legend = True, cmap = "GnBu", legend_kwds={'shrink': 0.3})
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_title(f'Average Violation Points in each city in each {year}', size = 15, weight = 'bold')
```

```python
for Season in Seasons:
    df_Violation_Points = df[['Name', 'City', 'Violation_Points', 'Season']].copy()
    df_Violation_Points = df_Violation_Points[df_Violation_Points['Season'] == Season]
    df_Violation_Points = df_Violation_Points.groupby('City').mean('Violation_Points').round()
    shape_Violation_Points = pd.merge(left=shape, right=df_Violation_Points, left_on="NAME", right_on="City", how = 'inner')
    shape_Violation_Points = gpd.GeoDataFrame(shape_Violation_Points)
    ax = shape_Violation_Points.boundary.plot(edgecolor = 'black', linewidth = 0.2, figsize = (8,8))
    shape_Violation_Points.plot(ax = ax , column ='Violation_Points',legend = True, cmap = "GnBu", legend_kwds={'shrink': 0.3})
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_title(f'Average Violation Points in each city in {Season}', size = 15, weight = 'bold')
```

```python
for Month in Months:
    df_Violation_Points = df[['Name', 'City', 'Violation_Points', 'Month']].copy()
    df_Violation_Points = df_Violation_Points[df_Violation_Points['Month'] == Month]
    df_Violation_Points = df_Violation_Points.groupby('City').mean('Violation_Points').round()
    shape_Violation_Points = pd.merge(left=shape, right=df_Violation_Points, left_on="NAME", right_on="City", how = 'inner')
    shape_Violation_Points = gpd.GeoDataFrame(shape_Violation_Points)
    ax = shape_Violation_Points.boundary.plot(edgecolor = 'black', linewidth = 0.2, figsize = (8,8))
    shape_Violation_Points.plot(ax = ax , column ='Violation_Points',legend = True, cmap = "GnBu", legend_kwds={'shrink': 0.3})
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_title(f'Average Violation Points in each city in {Month}', size = 15, weight = 'bold')
```

```python
for day_of_week in day_of_weeks:
    df_Violation_Points = df[['Name', 'City', 'Violation_Points', 'day_of_week']].copy()
    df_Violation_Points = df_Violation_Points[df_Violation_Points['day_of_week'] == day_of_week]
    df_Violation_Points = df_Violation_Points.groupby('City').mean('Violation_Points').round()
    shape_Violation_Points = pd.merge(left=shape, right=df_Violation_Points, left_on="NAME", right_on="City", how = 'inner')
    shape_Violation_Points = gpd.GeoDataFrame(shape_Violation_Points)
    ax = shape_Violation_Points.boundary.plot(edgecolor = 'black', linewidth = 0.2, figsize = (8,8))
    shape_Violation_Points.plot(ax = ax , column ='Violation_Points',legend = True, cmap = "GnBu", legend_kwds={'shrink': 0.3})
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_title(f'Average Violation Points in each city in {day_of_week}', size = 15, weight = 'bold')
```
