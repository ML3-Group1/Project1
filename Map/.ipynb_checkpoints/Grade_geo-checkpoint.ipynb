```python
import numpy as np
import pandas as pd
import shapefile as shp
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import plotly.express as px
from geopy.geocoders import Nominatim
from matplotlib.animation import FuncAnimation, PillowWriter
import calendar
```

```python
df = pd.read_csv('Clean_Data.csv')
df.drop("Unnamed: 0",axis =1,inplace=True)
df['Month'] = df['Month'].apply(lambda x: calendar.month_abbr[x])
df["City"] = df["City"].str.lower()
```

```python
shape = gpd.read_file("kx-king-county-wa-cities-and-unincorporated-areas-SHP/king-county-wa-cities-and-unincorporated-areas.shp")
shape["NAME"] = shape["NAME"].str.lower()
shape["NAME"].unique()
shape = shape.drop(labels=["UNINC"],axis = 1)
shape.boundary.plot()
```

```python
df_Grade = df[['Name', 'City', 'Grade']].copy()
df_Grade = df_Grade.groupby('City').mean('Grade').round()
shape_Grade = pd.merge(left=shape, right=df_Grade, left_on="NAME", right_on="City", how = 'inner')
shape_Grade = gpd.GeoDataFrame(shape_Grade)
ax = shape_Grade.boundary.plot(edgecolor = 'black', linewidth = 0.2, figsize = (8,8))
shape_Grade.plot(ax = ax , column ='Grade',legend = True, cmap = "GnBu", legend_kwds={'shrink': 0.3})
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.set_title('Average Grade in each city', size = 15, weight = 'bold')
```

```python
Years = df["Year"].unique().tolist()
Seasons = df["Season"].unique().tolist()
Months = df["Month"].unique().tolist()
day_of_weeks = df["day_of_week"].unique().tolist()
```

```python
for year in Years:
    df_Grade = df[['Name', 'City', 'Grade','Year']].copy()
    df_Grade = df_Grade[df_Grade['Year'] == year]
    df_Grade = df_Grade.groupby('City').mean('Grade').round()
    shape_Grade = pd.merge(left=shape, right=df_Grade, left_on="NAME", right_on="City", how = 'inner')
    shape_Grade = gpd.GeoDataFrame(shape_Grade)
    ax = shape_Grade.boundary.plot(edgecolor = 'black', linewidth = 0.2, figsize = (8,8))
    shape_Grade.plot(ax = ax , column ='Grade',legend = True, cmap = "GnBu", legend_kwds={'shrink': 0.3})
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_title(f'Average Grade in each city in year {year}', size = 15, weight = 'bold')
```

```python
for Season in Seasons:
    df_Grade = df[['Name', 'City', 'Grade','Season']].copy()
    df_Grade = df_Grade[df_Grade['Season'] == Season]
    df_Grade = df_Grade.groupby('City').mean('Grade').round()
    shape_Grade = pd.merge(left=shape, right=df_Grade, left_on="NAME", right_on="City", how = 'inner')
    shape_Grade = gpd.GeoDataFrame(shape_Grade)
    ax = shape_Grade.boundary.plot(edgecolor = 'black', linewidth = 0.2, figsize = (8,8))
    shape_Grade.plot(ax = ax , column ='Grade',legend = True, cmap = "GnBu", legend_kwds={'shrink': 0.3})
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_title(f'Average Grade in each city in {Season}', size = 15, weight = 'bold')
```

```python
for Month in Months:
    df_Grade = df[['Name', 'City', 'Grade','Month']].copy()
    df_Grade = df_Grade[df_Grade['Month'] == Month]
    df_Grade = df_Grade.groupby('City').mean('Grade').round()
    shape_Grade = pd.merge(left=shape, right=df_Grade, left_on="NAME", right_on="City", how = 'inner')
    shape_Grade = gpd.GeoDataFrame(shape_Grade)
    ax = shape_Grade.boundary.plot(edgecolor = 'black', linewidth = 0.2, figsize = (8,8))
    shape_Grade.plot(ax = ax , column ='Grade',legend = True, cmap = "GnBu", legend_kwds={'shrink': 0.3})
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_title(f'Average Grade in each city in {Month}', size = 15, weight = 'bold')
```

```python
for day_of_week in day_of_weeks:
    df_Grade = df[['Name', 'City', 'Grade','day_of_week']].copy()
    df_Grade = df_Grade[df_Grade['day_of_week'] == day_of_week]
    df_Grade = df_Grade.groupby('City').mean('Grade').round()
    shape_Grade = pd.merge(left=shape, right=df_Grade, left_on="NAME", right_on="City", how = 'inner')
    shape_Grade = gpd.GeoDataFrame(shape_Grade)
    ax = shape_Grade.boundary.plot(edgecolor = 'black', linewidth = 0.2, figsize = (8,8))
    shape_Grade.plot(ax = ax , column ='Grade',legend = True, cmap = "GnBu", legend_kwds={'shrink': 0.3})
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_title(f'Average Grade in each city in {day_of_week}', size = 15, weight = 'bold')
```
