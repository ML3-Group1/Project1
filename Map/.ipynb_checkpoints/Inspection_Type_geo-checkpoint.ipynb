```python
import numpy as np
import pandas as pd
import shapefile as shp
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import plotly.express as px
from geopy.geocoders import Nominatim
from matplotlib.animation import FuncAnimation, PillowWriter
import calendar
```

```python
df = pd.read_csv('Clean_Data.csv')
df.drop("Unnamed: 0",axis =1,inplace=True)
df['Month'] = df['Month'].apply(lambda x: calendar.month_abbr[x])
df["City"] = df["City"].str.lower()
```

```python
shape = gpd.read_file("kx-king-county-wa-cities-and-unincorporated-areas-SHP/king-county-wa-cities-and-unincorporated-areas.shp")
shape["NAME"] = shape["NAME"].str.lower()
shape["NAME"].unique()
shape = shape.drop(labels=["UNINC"],axis = 1)
shape.boundary.plot()
```

```python
Years = df["Year"].unique().tolist()
Seasons = df["Season"].unique().tolist()
Months = df["Month"].unique().tolist()
day_of_weeks = df["day_of_week"].unique().tolist()
```

```python
df_Inspection_Type = df[['Name', 'City', 'Inspection_Type']].copy()
df_Inspection_Type= df_Inspection_Type.groupby(['City',"Inspection_Type"],as_index=False).size().groupby("City").max()
shape_Inspection_Type = pd.merge(left=shape, right=df_Inspection_Type, left_on="NAME", right_on="City", how = 'inner')
shape_Inspection_Type = gpd.GeoDataFrame(shape_Inspection_Type)
ax = shape_Inspection_Type.boundary.plot(edgecolor = 'black', linewidth = 0.2, figsize = (8,8))
shape_Inspection_Type.plot(ax = ax , column ='size',legend = True, cmap = "GnBu", legend_kwds={'shrink': 0.3})
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.set_title('Number of routine inceptions in each city', size = 15, weight = 'bold')
```

```python
for year in Years:
    df_Inspection_Score = df[['Name', 'City', 'Inspection_Score','Year']].copy()
    df_Inspection_Score =  df_Inspection_Score[ df_Inspection_Score['Year'] == year]
    df_Inspection_Score = df_Inspection_Score.groupby('City').mean('Inspection_Score').round()
    shape_Inspection_Score = pd.merge(left=shape, right=df_Inspection_Score, left_on="NAME", right_on="City", how = 'inner')
    shape_Inspection_Score = gpd.GeoDataFrame(shape_Inspection_Score)
    ax = shape_Inspection_Score.boundary.plot(edgecolor = 'black', linewidth = 0.2, figsize = (8,8))
    shape_Inspection_Score.plot(ax = ax , column ='Inspection_Score',legend = True, cmap = "GnBu", legend_kwds={'shrink': 0.3})
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_title(f'Average Inspection Score in each city in {year}', size = 15, weight = 'bold')
```
