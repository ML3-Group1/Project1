```python
import numpy as np
import pandas as pd
import shapefile as shp
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import plotly.express as px
from geopy.geocoders import Nominatim
from matplotlib.animation import FuncAnimation, PillowWriter
import calendar
```

```python
df = pd.read_csv('Clean_Data.csv')
df.drop("Unnamed: 0",axis =1,inplace=True)
df['Month'] = df['Month'].apply(lambda x: calendar.month_abbr[x])
df["City"] = df["City"].str.lower()
```

```python
shape = gpd.read_file("kx-king-county-wa-cities-and-unincorporated-areas-SHP/king-county-wa-cities-and-unincorporated-areas.shp")
shape["NAME"] = shape["NAME"].str.lower()
shape["NAME"].unique()
shape = shape.drop(labels=["UNINC"],axis = 1)
shape.boundary.plot()
```

```python
Years = df["Year"].unique().tolist()
Seasons = df["Season"].unique().tolist()
Months = df["Month"].unique().tolist()
day_of_weeks = df["day_of_week"].unique().tolist()
```

```python
df_Inspection_Result = df[['Name', 'City', 'Inspection_Result']].copy()
df_Inspection_Result= df_Inspection_Result.groupby(['City',"Inspection_Result"],as_index=False).size().groupby("City").max()
shape_Inspection_Result = pd.merge(left=shape, right=df_Inspection_Result, left_on="NAME", right_on="City", how = 'inner')
shape_Inspection_Result = gpd.GeoDataFrame(shape_Inspection_Result)
ax = shape_Inspection_Result.plot(edgecolor = 'black', linewidth = 0.2, figsize = (8,8))
shape_Inspection_Result.plot(ax = ax , categorical=True,column ='Inspection_Result',legend = True, cmap = "GnBu")
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.set_title('Most inception results in each city', size = 15, weight = 'bold')
```

```python
for year in Years:
    df_Inspection_Result = df[['Name', 'City', 'Inspection_Result','Year']].copy()
    df_Inspection_Result =  df_Inspection_Result[ df_Inspection_Result['Year'] == year]
    df_Inspection_Result= df_Inspection_Result.groupby(['City',"Inspection_Result"],as_index=False).size().groupby("City").max()
    shape_Inspection_Result = pd.merge(left=shape, right=df_Inspection_Result, left_on="NAME", right_on="City", how = 'inner')
    shape_Inspection_Result = gpd.GeoDataFrame(shape_Inspection_Result)
    ax = shape_Inspection_Result.plot(edgecolor = 'black', linewidth = 0.2, figsize = (8,8))
    shape_Inspection_Result.plot(ax = ax , categorical=True,column ='Inspection_Result',legend = True, cmap = "GnBu")
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.set_title(f'Most inception results in each city in {year}', size = 15, weight = 'bold')
```
