# مرحله یک:کتابخانه های مربوطه

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#  مرحله دوم:اسم گذاری صحیح ستون ها در صورت نیاز


```python
df1=pd.read_csv('Food_Establishment_Inspection_Data(1).csv')
df1.head()
```

# مرحله سوم
مشخص کردن تایپ ستون ها                
و دسته بندی ستون ها                
و مشخص کردن ستونهای categorical

```python
df1.dtypes
```

# مرحله چهارم: حذف ستون های بدون کاربرد
'Program Identifier',  'Address',   'Phone'  ,'Business_ID'  ,'Violation_Record_ID'   ,'Zip Code'   ,'Inspection_Serial_Num'

```python
#Dropping unusable columns
df2=df1.drop(['Program Identifier','Address','Phone','Business_ID','Violation_Record_ID','Zip Code'],axis='columns')
df2.head(20)
```

# مرحله پنجم
پیدا کردن مقدار نول های ستون ها

```python
df2.isna().sum()
```

# مرحله ششم:
پاکسازی سطرهایی که بیشتر از سه مورد در ستونه انها مورد نول وجود دارد چون در این موارد هیچ اطلاعاتی نمیتوان به دست اورد

```python
#Find the rows with more than three columns with Nan values
df3=df2[df2.isna().sum(axis=1)>3]
df3.head(5)
```

```python
#clean the rows with more than two columns with Nan values
df3 = df2.dropna( how='all',
                          subset=['Inspection Business Name','Inspection Type', 'Violation Type','Inspection Score','Inspection Result','Inspection Closed Business','Violation Type'])
```

# بعد از مرحله ششم تعداد ستون هایی که دارای نول هستند بسیار کم شد و در مرحله بعد به پر کردن انها میپردازیم

```python
df3.isna().sum()
```

# **مرحله هفتم** : پرکردن موردهای نول در ستون های مربوطه


# 1.FILL INSPECTON SCORE COLUMN
1.Inspection Result     
2.Inspection Score                
در این جا یک بررسی در مورد ستون های 1 و2 داشتیم تا یه دید کلی به دست بیاد
به این صورت که روی ستون 1 بر اساس ستون 2 یک میانگین گرفته شده تا یک امتیاز بندی براساس میزان رضایت داشته باشیم              

```python
#INSPECTION SCORE:the type of its is float so we should fill it with numbers
df_mean=np.round(df3.groupby(['Inspection Result']).agg({'Inspection Score':'mean'}))
    
feature_data = df_mean.index
feature_count=df_mean['Inspection Score']

# visualize
plt.figure(figsize = (20,3))
plt.bar(feature_data,feature_count)
plt.xticks(feature_count.index, feature_count.index.values,rotation=45)
plt.ylabel("Frequency")
plt.title('Inspection Score')
plt.show()
```

در این جا مورد های satisfactory , complete هم دارای امتیاز تخلف هستند بنابراین نمیتوان مقدار انها را صرفا صفر در نظر گرفت

با توجه به حدود ۴۰ رکوردی که فاقد امتیاز بود، و حجم نمونه که بالای ۱۱۰۰۰۰ بود، که تعداد رکورد هایی که خالی بودند، کمتر از ۰.۱ درصد هستند پر کردن، آنها را با مقدار میانگین (بدون در نظر گرفتن واریانس آن) جایگزین کنیم و فیچر مورد نظر جهت دار نشود

# پیدا کردن ایندکس های دارای مورد نول

```python
idx1=df3[df3["Inspection Score"].isnull()].index
idx1
```

در این جا ستون 2 بر اساس میانگین نمرات طبق ستون 1 منظور نمرات بر اساس میزان رضایت است پر شده 

```python
df3['Inspection Score'] =np.round( df3.groupby(['Inspection Result'])['Inspection Score']
                     .apply(lambda x: x.fillna(x.mean())))
```

```python
df3.loc[idx1].head()
```

#2.FILL VIOLATION TYPE COLUMN
1.Violation type      
2.Inspection score        
در این جا بررسی شده که مجموع امتیاز های نارضایتی چه نوع کارتی را بیشتر دریافت کردند یعنی بیشتر رستوران ها چه نوع کارتی دریافت کردند در این تحلیل بیشتر رستوران ها دارای کارت قرمز هستند

```python
#violation type Nan values
idx2=df3[df3["Violation Type"].isnull()].index
df_mean=np.round(df3.groupby(['Violation Type']).agg({'Inspection Score':'mean'}))
    
feature_data = df_mean.index
feature_count=df_mean['Inspection Score']

# visualize
plt.figure(figsize = (20,3))
plt.bar(feature_data,feature_count)
plt.xticks(feature_count.index, feature_count.index.values,rotation=45)
plt.ylabel("Frequency")
plt.title('Violation Type')
plt.show()
```

```python
idx2=df3[df3["Violation Type"].isnull()].index
idx2
```

# در این مرحله ستون 1 را با _ پر کردم
به نظرم رستوران هایی که ستون 1 ان ها با نول پر شده یعنی هیچ تخلفی نداشتند و هیچ کارتی دریافت نکردند

```python
df3['Violation Type'] = df3['Violation Type'].fillna("_")
df3.loc[idx2].head(2)
```

#3.FILL VIOLATION DESCRIPTION
1.Violation Description         
2.Inspection Score            
 در این جا بررسی شده کدام مورد های تخلف میانگین امتیاز نارضایتی بیشتری دریاف کرده اند

```python
#violation Description
idx3=df3[df3["Violation Description"].isnull()].index
df_mean=np.round(df3.groupby(['Violation Description']).agg({'Inspection Score':'mean'}))
    
# get feature
feature_data = df_mean.index
# count number of data(value/sample)
feature_count=df_mean['Inspection Score']

# visualize
plt.figure(figsize = (22,3))
plt.bar(feature_data,feature_count)
plt.xticks(feature_count.index, feature_count.index.values,rotation=90)
plt.ylabel("Frequency")
plt.title('Violation Description')
plt.show()
```

و نتیجه این که ستون 1 که دارای مقدار نول هستند هیچ تخلفی انجام نداده اند و با مقدار _ پر میشوند

```python
#in this part we fill violation description with '_'
df3['Violation Description'] = df3['Violation Description'].fillna("_")
df3.loc[idx3].head(2)
```

#4.FILL GRADE COLUMN               
در اینجا بررسی شده که کدام درجه دارای امتیاز نارضایتی بیشتر است

```python
#GRADE
#GRADE
df_mean=np.round(df3.groupby(['Grade']).agg({'Inspection Score':'mean'}))
    
# get feature
feature_data = df_mean.index
# count number of data(value/sample)
feature_count=df_mean['Inspection Score']

# visualize
plt.figure(figsize = (20,3))
plt.bar(feature_data,feature_count)
plt.xticks(feature_count.index, feature_count.index.values,rotation=90)
plt.ylabel("Frequency")
plt.title('Inspection Score')
plt.show()
```

```python
idx=df3[df3["Grade"].isnull()].index
idx
```

در اینجا بر اساس میانگین های امتیاز تخلف ستون گرید پر شده است

```python
df3['Grade'] =np.round( df3.groupby(['Inspection Score'])['Grade']
                     .apply(lambda x: x.fillna(x.mean())))
df3.loc[idx].head(2)
```

#5.FILL LONGTITUDE AND LATITUDE COLUMNS                


در اینجا برای پر کردن مقدار نول ستون های طول و عرض جغرافیایی از متد استفاده شده است interpolate


```python
idx=df3[df3["Longitude"].isnull()].index
idx
```

```python
df3["Longitude"]=df3["Longitude"].interpolate(method='linear')
df3.loc[idx].head(2)
```

```python
df3["Latitude"]=df3["Latitude"].interpolate(method='linear')
df3.loc[idx].head(2)
```
