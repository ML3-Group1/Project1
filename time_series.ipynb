```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
from datetime import datetime
import plotly.express as px
```

```python
df1=pd.read_csv('final_data1.csv')
df1.head()
```

```python
df1['Inspection_Date'] = pd.to_datetime(df1['Inspection_Date'])
df1.head()
```

```python
df1['Year'] = pd.DatetimeIndex(df1['Inspection_Date']).year
df1['Month'] = pd.DatetimeIndex(df1['Inspection_Date']).month
df1['day_of_week'] = df1['Inspection_Date'].dt.day_name()
df1.head()
```

```python
def find_season(month, hemisphere):
    if hemisphere == 'Southern':
        season_month_south = {
            12:'Summer', 1:'Summer', 2:'Summer',
            3:'Autumn', 4:'Autumn', 5:'Autumn',
            6:'Winter', 7:'Winter', 8:'Winter',
            9:'Spring', 10:'Spring', 11:'Spring'}
        return season_month_south.get(month)
        
    elif hemisphere == 'Northern':
        season_month_north = {
            12:'Winter', 1:'Winter', 2:'Winter',
            3:'Spring', 4:'Spring', 5:'Spring',
            6:'Summer', 7:'Summer', 8:'Summer',
            9:'Autumn', 10:'Autumn', 11:'Autumn'}
        return season_month_north.get(month)
    else:
        print('Invalid selection. Please select a hemisphere and try again')
```

```python

season_list = []
hemisphere = 'Northern'
for month in df1['Month']:
    season = find_season(month, hemisphere)
    season_list.append(season)
    
df1['Season'] = season_list
df1.head()
```
